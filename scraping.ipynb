```python
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()

driver.get("https://www.traveloka.com/en-id/hotel/search?spec=25-05-2024.26-05-2024.1.1.HOTEL_GEO.102760.Kuta.1")
```

```python
panjang = 45
awal = 20
```

```python
dict_hotel = {}

for i in range(awal, panjang, 1):
    for j in range(1, 4, 1):
        try :
            xpath_hotel  = f"/html/body/div[1]/div[5]/div[2]/div/div[2]/div[5]/div/div[{i}]/div/div/div/div[2]/div[1]/div[1]/div[{j}]/h3"
            dict_hotel[i] = driver.find_elements(By.XPATH, xpath_hotel)[0].text
            print(i, dict_hotel[i])
        except:
            continue
```

```python
dict_harga = {}

for i in range(awal, panjang, 1): 
    try :
        xpath_harga = f"/html/body/div[1]/div[5]/div[2]/div/div[2]/div[5]/div/div[{i}]/div/div/div/div[3]/div/div/div[1]/div[3]/div/div[3]/div[1]"
        dict_harga[i] = driver.find_elements(By.XPATH, xpath_harga)[0].text
        print(i, dict_harga[i])
    except IndexError:
        continue
```

```python
dict_reviews = {}

for i in range(awal, panjang, 1):
    for j in range(1, 4, 1):
        try :
            xpath_reviews = f"/html/body/div[1]/div[5]/div[2]/div/div[2]/div[5]/div/div[{i}]/div/div/div/div[2]/div[1]/div[1]/div[{j}]/div[1]/div[2]/span"
            dict_reviews[i] = driver.find_elements(By.XPATH, xpath_reviews)[0].text
            print(i, dict_reviews[i])
        except IndexError:
            continue
```

```python
dict_rating = {}

for i in range(awal, panjang, 1):
    for j in range(1, 4, 1):
        try :
            xpath_rating = f"/html/body/div[1]/div[5]/div[2]/div/div[2]/div[5]/div/div[{i}]/div/div/div/div[2]/div[1]/div[1]/div[{j}]/div[1]/div[2]"
            dict_rating[i] = driver.find_elements(By.XPATH, xpath_rating)[0].text
            print(i, dict_rating[i])
        except IndexError:
            continue
```

```python
dict_facilities = {}

for i in range(awal, panjang, 1): 
    try :
        xpath_facilities = f"/html/body/div[1]/div[5]/div[2]/div/div[2]/div[5]/div/div[{i}]/div/div/div/div[2]/div[1]/div[2]/div[3]/div[2]"
        dict_facilities[i] = driver.find_elements(By.XPATH, xpath_facilities)[0].text
        print(i, f"{dict_facilities[i]}")
    except IndexError:
        continue
```

```python
nama = []
harga = []
reviews = []
rating = []
facilities = []

# buatkan fungsi untuk menghitung panjang minimum dari dictionary, dan kembalikan dictionary
def min_length():
    return min(len(dict_hotel), len(dict_harga), len(dict_reviews), len(dict_rating), len(dict_facilities))

def dict_minimum():
    n = min_length()

    if(len(dict_hotel) == n):
        return dict_hotel
    elif(len(dict_harga) == n):
        return dict_harga
    elif(len(dict_reviews) == n):
        return dict_reviews
    elif(len(dict_rating) == n):
        return dict_rating
    else:
        return dict_facilities

dict_iterator = dict_minimum()

for i in dict_iterator.keys():
    try :
        nama.append(dict_hotel[i])
        harga.append(dict_harga[i])
        reviews.append(dict_reviews[i])
        rating.append(dict_rating[i])
        facilities.append(dict_facilities[i])
    except KeyError:
        continue
    
print(len(nama))
print(len(harga))
print(len(reviews))
print(len(rating))
print(len(facilities))
```

```python
import pandas as pd

df = pd.DataFrame({'nama': nama, 'harga': harga, 'reviews': reviews, 'rating': rating, 'facilities': facilities})
df
```

```python
df.to_csv("part15.csv", index=False)
```
